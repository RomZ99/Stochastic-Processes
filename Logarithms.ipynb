# Why linear approximation function is a decent substitute for a function(x) at a particular point?(compares the function f(x) with the approximation of its linear function l(x) around a point a)

for very small x ln(1+x) ~ x

```python
def lin_approx(f,df,a):
    def l(x) :
        return  f(a)+ (df(a)*(x-a))
    return l    

f = lambda x: np.log(1 + x)
df = lambda x: 1/(1+x)
a=1

L= lin_approx(f,df,a)

r= [0.99,1,1.001,1.01,1.05,2,3,4,4.0001,4.001,4.01,4.02,4.04,5]

results= pd.DataFrame({'x': r,'ln(1+x)':[f(x) for x in r],'l(x)':[L(x) for x in r]})
results['Error'] = (results['ln(1+x)'] - results['l(x)']).round(5)
results
```
